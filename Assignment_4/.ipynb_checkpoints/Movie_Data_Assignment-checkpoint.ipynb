### **Question 4**
Which movies were released after the year 2000? (Year 2000 ke baad release hone wali movies filter karein).

```python
DataSet['Released_Year'] = pd.to_numeric(DataSet['Released_Year'], errors='coerce')
# DataSet.Released_Year.isna().sum() 
movies_after_2000 = DataSet[DataSet['Released_Year'] > 2000].copy()
movies_after_2000['Released_Year'] = movies_after_2000['Released_Year'].astype(int)
movies_after_2000.head()
```

### **Question 11**
Which movie has the longest runtime? (Sabse lambi movie ka naam aur runtime minutes mein nikalye).

```python
# 1. Pehle ' min' text ko hata kr number ma convert kr raha hun

DataSet['Runtime_Num'] = DataSet['Runtime'].str.replace(' min', '').astype(int)

# 2. PHir .max() lagya hai so that subse ziada duration samne ajae

longest_movie = DataSet[DataSet['Runtime_Num'] == DataSet['Runtime_Num'].max()]

# 3. phir isko new variable ma store kr k display krwa raha hun

longest_movie[['Series_Title', 'Runtime']]
```

### **Question 16**
'Gross' column mein jo missing values (NaN) hain, unko pooray column ke median value se fill (impute) karein.

```python
gross_median = DataSet['Gross'].median()

DataSet['Gross'] = DataSet['Gross'].fillna(gross_median)

DataSet['Gross'].isna().sum()
```
