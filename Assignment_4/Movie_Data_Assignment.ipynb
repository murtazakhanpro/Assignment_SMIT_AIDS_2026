### **Question 36**
(Visualization) 'Drama' aur 'Action' genres ki movies ke 'Gross' revenue ko compare karne ke liye ek Grouped Box Plot ya Violin Plot banayein.

```python
# Pehle sirf 'Drama' aur 'Action' wale row chunks filter karein
drama_data = DataSet[DataSet['Genre'].str.contains('Drama', na=False)].copy()
drama_data['Genre_Category'] = 'Drama'

action_data = DataSet[DataSet['Genre'].str.contains('Action', na=False)].copy()
action_data['Genre_Category'] = 'Action'

# Dono filtered dataframes ko aapas mein jod (combine) dein
combined_genre_df = pd.concat([drama_data, action_data])

# Box Plot ke zariye Gross revenue ka muqabla (comparison) karein
plt.figure(figsize=(10, 6))
sns.boxplot(data=combined_genre_df, x='Genre_Category', y='Gross', palette='Set2')
plt.title('Gross Revenue Comparison: Drama vs Action')
plt.xlabel('Genre Category')
plt.ylabel('Gross Revenue')
plt.show()
```

### **Question 40**
(Visualization) Top 5 highest rated movies aur unke 'Meta_score' vs 'IMDB_Rating' (scaled to 100) ka ek Side-by-Side Comparison Bar Chart banayein.

```python
# Top 5 highest rated movies select karein
top_5_movies = DataSet.sort_values(by='IMDB_Rating', ascending=False).head(5)

# Scaled IMDB rating (jo humne Q19 mein banaya tha) aur Meta score ko side-by-side format mein laane ke liye reshape (.melt) karein
comparison_df = pd.DataFrame({
    'Movie': top_5_movies['Series_Title'],
    'IMDB Rating (x10)': top_5_movies['Rating_100_Scale'],
    'Meta Score': top_5_movies['Meta_score']
}).melt(id_vars='Movie', var_name='Rating Type', value_name='Score')

# Side-by-Side Bar Chart plot karein
plt.figure(figsize=(12, 6))
sns.barplot(data=comparison_df, x='Movie', y='Score', hue='Rating Type', palette='Set2')
plt.title('Comparison of IMDB Rating (Scaled) vs Meta Score for Top 5 Movies', fontsize=14, fontweight='bold')
plt.xticks(rotation=15, ha='right')
plt.xlabel('Movie Title')
plt.ylabel('Score (out of 100)')
plt.legend(title='Rating Type')
plt.show()
```
