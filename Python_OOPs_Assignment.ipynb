**Q4:** Protected variable (`_`) aur Private variable (`__`) ke beech kya farq hota hai? Aik simple code ke zariye dono ka use-case dikhayein.

```python
# Write your code here for Q4
# protected variable main eik underscore hota ha (_)
# private variable main do underscore hote hain (__)

class answer:
    def __init__(self, protected , private):
        self._protected = protected
        self.__private = private

    def protected(self):
        print("yeh private variable hai")
    def private(self):
        print("yeh private variable hai")

obj = answer("Main Protected hoon", "Main Private hoon")

print(obj._protected)
print(obj.__private)
```

**Q11:** Method Overriding kya hai? Parent class `Bird` banayein jisme `fly()` method ho jo print kare 'Most birds can fly'. Child class `Ostrich` banayein jo is method ko override kare aur print kare 'Ostriches cannot fly'.

```python
# 1. Parent Class banayi
class Bird:
    def fly(self):
        print("Most birds can fly")

# 2. Child Class banayi jis ne parent ke method ko override (tabdeel) kiya
class Ostrich(Bird):
    def fly(self):  # Same naam ka function dobara likha
        print("Ostriches cannot fly")

# --- Testing ---

# Bird ka object parent wala method chalaye ga
chiriya = Bird()
chiriya.fly()  # Output: Most birds can fly

# Ostrich ka object apna (overridden) method chalaye ga
shuturmurg = Ostrich()
shuturmurg.fly()  # Output: Ostriches cannot fly
```

**Q12:** Polymorphism function ke zariye kaise achieve hoti hai? Aik function banayein `make_sound(animal_object)` jo har tarah ke animal class ka sound method automatically call kare.

```python
# Write your code here for Q12
# 1. Dog Class banayi
class Dog:
    def sound(self):
        print("Woof! Woof!")

# 2. Cat Class banayi (is mein bhi sound naam ka function hai)
class Cat:
    def sound(self):
        print("Meow! Meow!")

# 3. Main Polymorphic Function banayan jo kisi bhi animal ka object le sakta hai
def make_sound(animal_object):
    # Yeh line automatically pass kiye gaye object ka sound() method chalaye gi
    animal_object.sound()

# --- Testing ---

# Alag alag classes ke objects banaye
moti = Dog()
mano = Cat()

# Aik hi function ko alag alag objects diye
make_sound(moti)  # Output: Woof! Woof!
make_sound(mano)  # Output: Meow! Meow!
```

**Q13:** Operator Overloading kya hoti hai? Do objects ko `+` operator ke zariye add karne ke liye `__add__` magic method ka use kaise karte hain? Code likhein.

```python
# Write your code here for Q13
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    # __add__ magic method jo '+' sign lagane par automatic chalta hai
    def __add__(self, other):
        # self matlab pehla object, other matlab doosra object
        new_x = self.x + other.x
        new_y = self.y + other.y
        # Plus karne ke baad hum aik naya object return karte hain
        return Point(new_x, new_y)

    # Yeh method output ko saaf suthra dekhane ke liye hota hai
    def __str__(self):
        return f"({self.x}, {self.y})"

# --- Testing ---

# Do alag alag objects banaye
p1 = Point(2, 4)
p2 = Point(3, 5)

# Ab hum dono ko '+' ke zariye plus kar sakte hain!
p3 = p1 + p2  # Backend par yeh p1.__add__(p2) ko chalata hai

print("P1:", p1)  # Output: (2, 4)
print("P2:", p2)  # Output: (3, 5)
print("P1 + P2 =", p3)  # Output: (5, 9) [2+3=5 aur 4+5=9]
```

**Q15:** Kya Python default roop se Method Overloading (same method name with different arguments) support karta hai? Agar nahi, toh hum same name ke method ko multiple arguments ke sath kaise handle karte hain? Code likhein.

```python
# Write your code here for Q15
class Calculator:
    # Hum arguments ko default value 'None' de dete hain
    def add(self, a=None, b=None, c=None):
        
        # Case 1: Agar teeno numbers diye gaye hon
        if a is not None and b is not None and c is not None:
            return a + b + c
            
        # Case 2: Agar sirf do numbers diye gaye hon
        elif a is not None and b is not None:
            return a + b
            
        # Case 3: Agar sirf aik number diya gaya ho
        elif a is not None:
            return a
            
        # Case 4: Agar kuch bhi na diya gaya ho
        else:
            return 0

# --- Testing ---

calc = Calculator()

# 1. Teen arguments ke sath chalaya
print("Teen numbers ka sum:", calc.add(5, 10, 15))  # Output: 30

# 2. Do arguments ke sath chalaya
print("Do numbers ka sum:", calc.add(10, 20))       # Output: 30

# 3. Aik argument ke sath chalaya
print("Sirf aik number:", calc.add(7))              # Output: 7

# 4. Bina kisi argument ke chalaya
print("Kuch bhi nahi diya:", calc.add())            # Output: 0
```

**Q17:** Q16 wali `Computer` class ko inherit karke do child classes `Laptop` aur `Desktop` banayein aur dono mein `process()` method ko mandatory implement karein.

```python
# Write your code here for Q17
from abc import ABC, abstractmethod

# 1. Parent Abstract Class (Q16 wali)
class Computer(ABC):
    @abstractmethod
    def process(self):
        pass

# 2. Child Class 1: Laptop
class Laptop(Computer):
    def process(self):  # Mandatory function implement kiya
        print("Laptop is processing with a portable battery.")

# 3. Child Class 2: Desktop
class Desktop(Computer):
    def process(self):  # Mandatory function implement kiya
        print("Desktop is processing with direct power supply.")

# --- Testing ---

# Ab in child classes ke objects bilkul sahi banenge!
my_laptop = Laptop()
my_desktop = Desktop()

# Dono ka apna apna process() method chalayein
my_laptop.process()   # Output: Laptop is processing...
my_desktop.process()  # Output: Desktop is processing...
```

**Q18:** Aik abstract class `Car` banayein jisme aik concrete (normal) method `start_engine()` ho aur aik abstract method `drive()` ho. Child class bana kar dono ko test karein.

```python
# Write your code here for Q18
from abc import ABC, abstractmethod

# 1. Abstract Class
class Car(ABC):
    
    # Concrete (Normal) Method - Is ka code pehle se tayyar hai
    def start_engine(self):
        print("Engine started... Vroom Vroom!")
        
    # Abstract Method - Yeh child class khud mukammal kare gi
    @abstractmethod
    def drive(self):
        pass

# 2. Child Class
class SportsCar(Car):
    # Abstract method ko implement karna lazmi hai
    def drive(self):
        print("Driving fast at 200 km/h!")

# --- Testing ---

# Child class ka object banaya
my_car = SportsCar()

# 1. Normal method chalaya (jo parent se direct mila)
my_car.start_engine()  # Output: Engine started... Vroom Vroom!

# 2. Abstract method chalaya (jo child ne khud likha)
my_car.drive()         # Output: Driving fast at 200 km/h!
```

**Q20:** Aik real-world scenario (jaise Payment System) design karein jahan `Payment` aik abstract class ho aur `CreditCardPayment` aur `UPIPayment` uski subclasses hon.

```python
# Write your code here for Q20
from abc import ABC, abstractmethod

# 1. Abstract Class - Jo poore system ka rule set karti hai
class Payment(ABC):
    @abstractmethod
    def process_payment(self, amount):
        pass

# 2. Subclass 1: Credit Card ke zariye payment
class CreditCardPayment(Payment):
    def __init__(self, card_number):
        self.card_number = card_number

    def process_payment(self, amount):
        print(f"Processing Credit Card payment of Rs.{amount}")
        print(f"Card ending in {self.card_number[-4:]} successfully charged!\n")

# 3. Subclass 2: UPI/Mobile Wallet ke zariye payment
class UPIPayment(Payment):
    def __init__(self, upi_id):
        self.upi_id = upi_id

    def process_payment(self, amount):
        print(f"Processing UPI payment of Rs.{amount}")
        print(f"Money deducted from UPI ID: {self.upi_id} successfully!\n")

# --- Testing ---

# User ne Credit Card select kiya
payment1 = CreditCardPayment("1234-5678-9876-5432")
payment1.process_payment(1500)

# User ne UPI select kiya
payment2 = UPIPayment("murtaza@ybl")
payment2.process_payment(500)
```
