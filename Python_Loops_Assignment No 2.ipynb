# **Q12:** User se ek number lein aur check karein ke wo **Prime Number** hai ya nahi (using loops).

````python
# type your code here
num = int(input("Please Enter A number:"))
is_prime = True
if num <= 1:
    is_prime = False

else:
    for i in range(2,num):
        if num % i==0:
            is_prime = False
            break

if is_prime:
    print(f'''{num} is a prime Numeber''')
else:
    print(f'''{num} is  not a prime Number''')


            
````

# **Q20:** Ek 'Number Guessing Game' banayein jahan program ek number (1-20) fix kare aur user ko 3 chances de guess karne ke liye. `while loop` aur `break` ka use karein.

````python
secret_number = 8
chances = 3 
attempt = 0
while attempt < chances:
    attempt = attempt+1
    guss_number = int(input("please enter a number between 1 to 20"))
    if guss_number == secret_number:
        print("Good job You Guessed the right number")
        break

    elif guss_number>= secret_number:
        print("Apka Number Bohat bada hai", guss_number)
    
    else:
        print("Apka guss Chota Hai", guss_number)


if guss_number != secret_number:
    print("Ghalat Number Try Again" , guss_number)
````
