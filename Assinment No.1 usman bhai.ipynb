### 43. Ek simple calculator banayein (+, -, *, /) symbols ke basis par operation perform karein.

```python
# Write your code here
# 1. User se inputs lein
num1 = float(input("Pehla number likhein: "))
operator = input("Operation chunain (+, -, *, /): ")
num2 = float(input("Doosra number likhein: "))

# 2. Conditions laga kar check karein
if operator == "+":
    print(f"Result: {num1 + num2}")
elif operator == "-":
    print(f"Result: {num1 - num2}")
elif operator == "*":
    print(f"Result: {num1 * num2}")
elif operator == "/":
    # Zero se divide karne ka check (Error handling)
    if num2 != 0:
        print(f"Result: {num1 / num2}")
    else:
        print("Error! Zero se divide nahi kar sakte.")
else:
    print("Ghalat symbol! Sirf +, -, *, / istemal karein.")
```

### 44. User ka weight aur height le kar BMI category batayein (Underweight, Normal, Overweight).

```python
# Write your code here
weight = float(input("Apna weight (kg) likhein: "))
height_feet = float(input("Apni height (feet) likhein: "))

# Feet ko meter mein badalne ke liye
height_meters = height_feet * 0.3048

# BMI Formula
bmi = weight / (height_meters * height_meters)

print("Aapka BMI hai:", round(bmi, 2))

if bmi < 18.5:
    print("Aap Underweight hain.")
elif bmi < 25:
    print("Aap Normal weight hain.")
else:
    print("Aap Overweight hain.")
```
