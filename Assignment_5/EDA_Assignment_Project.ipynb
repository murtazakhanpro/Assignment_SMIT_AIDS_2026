```python


fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# 1. Target Variable Count
sns.countplot(x='Loan_Status', data=df1, ax=axes[0], palette='Set2')
axes[0].set_title('Loan Approval Status (Target Variable)')

# 2. Applicant Income Distribution (Skewness check karne ke liye)
sns.histplot(df1['ApplicantIncome'], kde=True, ax=axes[1], color='purple')
axes[1].set_title('Applicant Income Distribution')

plt.tight_layout()
plt.show()

# Dependents aur Property Area ka Count plot
fig, axes = plt.subplots(1, 2, figsize=(16, 5))
sns.countplot(x='Dependents', data=df1, ax=axes[0], order=['0', '1', '2', '3+'])
sns.countplot(x='Property_Area', data=df1, ax=axes[1])
plt.show()
```
